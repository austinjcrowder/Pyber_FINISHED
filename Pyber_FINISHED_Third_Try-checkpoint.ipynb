## Total Fares by City Type

```python
# Calculate Percents
pie_chart_df = pd.DataFrame(rs_data.groupby("type")["fare"].sum())
pie_chart_df
# Build Pie Chart
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Urban", "Surburban", "Rural"]
explode = (0, 0, .1)
plt.pie(pie_chart_df['fare'], labels=labels, colors=colors,
        autopct="%1.1f%%", explode=explode, shadow=True, startangle=140)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("fares chart", dpi=180, bbox_inches="tight")
plt.show()
```

```python
# Show Figure
plt.show()
```
